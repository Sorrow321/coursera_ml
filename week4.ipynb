# Центральная предельная теорема 

```python
from scipy.stats import expon
from scipy.stats import norm
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

Используем экспоненциальное распределение с $\lambda = 1$(https://ru.wikipedia.org/wiki/Экспоненциальное_распределение) 

```python
r = expon.rvs(size=1000) # генерируем выборку
```

```python
x = np.linspace(0, 10, 100)
plt.hist(r, color = 'red', density=True, bins=20) # строим гистограмму
plt.plot(x, expon.pdf(x), color='blue') # строим график плотности показательного распределения
plt.xlabel('X')
plt.ylabel('плотность')
```

Смотрим на распределения выборочных средних $X_n$ для выборок объема $n$.<br><br>
По центральной предельной теореме$$X_n \approx ~ \mathrm{N}(\mathbb{E}X, \frac{\mathbb{D}X}{n})\\ n \rightarrow +\infty$$
Так как $X \sim \mathrm{Exp}(1)$, то $\mathbb{E}X = 1$, $\mathbb{D}X = 1$.<br>
Будем генерировать по 1000 раз выборки объема $n = 5, 10, 50$, брать их среднее выборочное и сравнивать результат с $\mathrm{N}(1, \frac{1}{n})$.

```python
n = 5
arr = np.empty(1)
for i in range(0, 1000): 
    k = expon.rvs(size=n) # наша выборка объема n
    arr = np.append(arr, np.sum(k) / n) # добавляем в массив выборочное среднее
plt.hist(arr, bins=50, density=True) # выводим гистограмму
x_axis = np.arange(0, 5, 0.001)
plt.plot(x_axis, norm.pdf(x_axis, 1, 1 / np.sqrt(n))) # выводим график плотности N(1, 1/n)
plt.xlabel('$X_5$')
plt.ylabel('плотность')
```

```python
n = 10
arr = np.empty(1)
for i in range(0, 1000): 
    k = expon.rvs(size=n) # наша выборка объема n
    arr = np.append(arr, np.sum(k) / n) # добавляем в массив выборочное среднее
plt.hist(arr, bins=50, density=True) # выводим гистограмму
x_axis = np.arange(0, 5, 0.001)
plt.plot(x_axis, norm.pdf(x_axis, 1, 1 / np.sqrt(n))) # выводим график плотности N(1, 1/n)
plt.xlabel('$X_{10}$')
plt.ylabel('плотность')
```

```python
n = 50
arr = np.empty(1)
for i in range(0, 1000):
    k = expon.rvs(size=n) # наша выборка объема n
    arr = np.append(arr, np.sum(k) / n) # добавляем в массив выборочное среднее
plt.hist(arr, bins=50, density=True) # выводим гистограмму
x_axis = np.arange(0, 4, 0.001)
plt.plot(x_axis, norm.pdf(x_axis, 1, 1 / np.sqrt(n))) # выводим график плотности N(1, 1/n)
plt.xlabel('$X_{50}$')
plt.ylabel('плотность')
```

Как мы видим, при увеличении объема выборки $n$ распределение выборочных средних стремится к соответствующему нормальному распределению, как и утверждает центральная предельная теорема.
