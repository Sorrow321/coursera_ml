```python
from sklearn.datasets import load_digits
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
loaded = load_digits()
```

```python
X = loaded.data
target = loaded.target
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.tree import DecisionTreeClassifier
```

```python
clf = DecisionTreeClassifier()
cv = cross_val_score(clf, X, target, cv=10)
print(cv.mean())
```

```python
from sklearn.ensemble import BaggingClassifier
```

```python
bclf = BaggingClassifier(n_estimators=100)
cv = cross_val_score(bclf, X, target, cv=10)
print(cv.mean())
```

```python
bclf = BaggingClassifier(max_features=8, n_estimators=100)
cv = cross_val_score(bclf, X, target, cv=10)
print(cv.mean())
```

```python
clf = DecisionTreeClassifier(max_features=8)
bclf = BaggingClassifier(base_estimator=clf, n_estimators=100)
cv = cross_val_score(bclf, X, target, cv=10)
print(cv.mean())
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=100, max_features=8)
cv = cross_val_score(rf, X, target, cv=10)
print(cv.mean())
```

```python
values = []
for i in range(1, 100, 5):
    rf = RandomForestClassifier(n_estimators=i, max_features=8)
    cv = cross_val_score(rf, X, target, cv=10)
    values.append(cv.mean())
```

```python
plt.plot(values)
```

```python
values = []
for i in range(1, 60, 5):
    rf = RandomForestClassifier(n_estimators=100, max_features=i)
    cv = cross_val_score(rf, X, target, cv=10)
    values.append(cv.mean())
```

```python
plt.plot(values)
```

```python
values = []
for i in range(1, 60, 5):
    rf = RandomForestClassifier(n_estimators=100, max_features=8, max_depth=i)
    cv = cross_val_score(rf, X, target, cv=10)
    values.append(cv.mean())
```

```python
plt.plot(values)
```
